## Data Cleaning & Preprocessing

In this section, we transform the raw dataset into a clean, structured, and analysis-ready format. Although the dataset is largely complete with no missing values, several preprocessing steps are required to ensure consistency and suitability for machine learning tasks.

The cleaning process includes:
- Standardizing column names for uniformity and compatibility  
- Converting binary categorical variables (Yes/No) into numeric format (0/1)  
- Verifying and visualizing missing values across all features  
- Identifying categorical variables for future encoding  
- Ensuring proper data types across all columns  
- Generating summary statistics for continuous numeric variables  

These steps are essential to eliminate inconsistencies, improve data quality, and prepare the dataset for downstream analysis such as exploratory data analysis and predictive modeling.

```python
# Standardize Column Names
# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)

print("✓ Cleaned column names:")
print(df.columns.tolist())
```

```python
# Convert binary categorical columns

# Automatically detect binary Yes/No columns
binary_cols = [
    col for col in df.columns
    if df[col].dtype == 'object'
    and df[col].dropna().isin(['Yes', 'No']).all()
]

# Convert to 0/1
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0}).astype(int)

print(f"✓ Converted {len(binary_cols)} binary columns")

# Show sample
df[binary_cols[:5]].head()
```

```python
# Compute missing values
missing_info = df.isna().sum()
missing_percent = (missing_info / len(df)) * 100

# Build summary table
missing_df = pd.DataFrame({
    'feature': missing_info.index,
    'missing_count': missing_info.values,
    'missing_percentage': missing_percent.values
})

# Sort by missing percentage
missing_df = missing_df.sort_values(
    by='missing_percentage',
    ascending=False
).reset_index(drop=True)

# Split into two halves
mid = int(np.ceil(len(missing_df) / 2))
left = missing_df.iloc[:mid].reset_index(drop=True)
right = missing_df.iloc[mid:].reset_index(drop=True)

# Align both halves
if len(left) > len(right):
    for _ in range(len(left) - len(right)):
        right = pd.concat([
            right,
            pd.DataFrame([{
                'feature': '',
                'missing_count': '',
                'missing_percentage': ''
            }])
        ], ignore_index=True)

# Combine side-by-side
combined_df = pd.DataFrame({
    'feature_left': left['feature'],
    'missing_count_left': left['missing_count'],
    'missing_percentage_left': left['missing_percentage'],
    'feature_right': right['feature'],
    'missing_count_right': right['missing_count'],
    'missing_percentage_right': right['missing_percentage'],
})

# Styling function
def highlight_missing(val):
    if val == '' or pd.isna(val):
        return ''
    elif val > 50:
        return 'background-color: red; color: white; font-weight: bold'
    elif val > 20:
        return 'background-color: orange; font-weight: bold'
    elif val > 5:
        return 'background-color: yellow'
    return ''

# Apply styling safely
styled_df = combined_df.style.apply(
    lambda col: col.map(highlight_missing),
    subset=[
        'missing_percentage_left',
        'missing_percentage_right'
    ]
)

styled_df
```

- We had no missing in our data
